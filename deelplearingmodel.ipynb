```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
```

```python
import tensorflow
tensorflow.keras.__version__
```

```python
df = pd.read_csv('student-mat-por-merge.csv',sep=",",header=0)
```

```python
df
```

```python
grade = []
for x in df["G3"]:

    if x > 15:
        grade.append(4)
    elif x >= 10:
        grade.append(3)
    elif x >= 5:
        grade.append(2)
    else:
        grade.append(1)
print(grade)
```

```python
df["grade"] = grade
df
```

```python
new_df = df.drop(columns = ["G2","G1","G3"])
new_df
```

```python
data = new_df.values
X = new_df.drop(columns =["grade"])
y = new_df['grade']
y
```

```python
cat_columns = ["school","sex","address","famsize","Pstatus","Mjob","Fjob","reason","guardian","schoolsup","famsup","paid","activities","nursery","higher","internet","romantic","subject"]
cat_columns2 = ["school","absences","age","health","Walc","goout","subject"]

cat_columns2
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
label_encoder = LabelEncoder()
```

```python
X= X[cat_columns2]
##X= X.apply(LabelEncoder().fit_transform)
X
```

```python
X[["school","subject"]] = X[["school","subject"]].apply(LabelEncoder().fit_transform)
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import sklearn
from tensorflow.keras.utils import to_categorical
```

```python
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=7))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))

model.add(Dense(units=5, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=190,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:10])
print(encoded_predictions)
```

```python
print(f'actual {y_test.head(10)} prediction{encoded_predictions[:10]}')
```
